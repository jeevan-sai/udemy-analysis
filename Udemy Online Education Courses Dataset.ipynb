```python
cos=[ 'price', 'num_subscribers', 'num_reviews', 'num_lectures','content_duration']
```

```python
for i in cos:
    print(i)
    skew=datacourse[i].skew().round()
    print(skew)
    print("----------------")
```

```python
for i in cos:
    sns.displot(datacourse[i])
    print(i)
    print("------")
```

```python
for i in cos:
    skew=datacourse[i].skew()
    if skew !=0:
        q1=datacourse[i].quantile(0.25)
        q3=datacourse[i].quantile(0.75)
        iqr=q3-q1
        up=q3+1.5*iqr
        lp=q3-1.5*iqr
        datacourse[i]=datacourse[i].map(lambda x : up if x>up else lp if x<lp else x)
    else :
        m=datacourse[i].mean()
        s=datacourse[i].std()
        ul=m+3*s
        ll=m-3*s
        datacourse[i]=datacourse[i].map(lambda x : ul if x>ul else ll if x<ll else x)
```
